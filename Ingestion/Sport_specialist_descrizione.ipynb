```python
#apro un file txt e ci incollo il testo dentro
file = open(r'C:\Users\Utente\Desktop\My_Share\sport_specialist\descrizione.txt', encoding = 'utf8')
file
type(file)
```

```python
#creo una lista in cui ogni riga del testo diventa stringa
lines=[]
for line in file.readlines():
    lines.append(line)
lines
```

```python
description = []
riga_correct = []


tabella = []

num = 0
write_description = False

for i, line in enumerate(lines):
    riga = line.replace('\t','').replace('\n','').split(',')
    if i == 0:
        #aggiungo la colonna name che contiene il nome del prodotto
        #riga[3] = 'name'
        #riga[4] = 'descrizione_prodotto'
        #riga.append(riga[-1])
        #riga[-2] = 'sostenibile'        
        #riga.append(riga[-1])
        del riga[-1]
        
    try:
        index = int(riga[0])
    except Exception as e:
        index = riga[0]
        
    if index == num:
        if write_description:
            riga_correct.append(description)
            tabella.append(riga_correct)
            
        riga_correct = riga
        description = []
        write_description = False
        num += 1
    else:
        write_description = True
        description.append(riga)
tabella
```
